### Clusters

```python
s_files = [os.path.join('/pscratch/sd/m/mjyb16/DR9/results_south_2', f) for f in os.listdir('/pscratch/sd/m/mjyb16/DR9/results_south_2') if os.path.isfile(os.path.join('/pscratch/sd/m/mjyb16/DR9/results_south_2', f))]
n_files = [os.path.join('/pscratch/sd/m/mjyb16/DR9/results_north_2', f) for f in os.listdir('/pscratch/sd/m/mjyb16/DR9/results_north_2') if os.path.isfile(os.path.join('/pscratch/sd/m/mjyb16/DR9/results_north_2', f))]
```

```python
list_of_tables = []
for file in n_files:
    t = Table.read(file)
    list_of_tables.append(t)
```
