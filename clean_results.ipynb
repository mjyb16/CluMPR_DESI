# Mass correction

```python
data = Table.read("all_clusters4.fits").to_pandas()
```

```python
data.drop(columns = ["ra_indices", "dec_indices"], inplace = True)
```

```python
data.edge_mask = data.edge_mask.astype("int")
```

```python
data
```

```python
data.iloc[0]
```

```python
def mass_coefficient(z):
    return 10**((-0.010062146327355921)*z**2 + 0.4931848214696567*z + 0.07716204385814744)
```

```python
plt.plot(data.z_median_central, (mass_coefficient(data.z_median_central)/data.correction_factor), ".")
```

```python
data["neighbor_mass"] = np.log10((10**data.neighbor_mass)*(mass_coefficient(data.z_median_central)/data.correction_factor))
```

```python
data["local_neighbor_mass"] = (10**data.local_neighbor_mass)*(mass_coefficient(data.z_median_central)/data.correction_factor)
```

```python
data["ultra_local_neighbor_mass"] = (10**data.ultra_local_neighbor_mass)*(mass_coefficient(data.z_median_central)/data.correction_factor)
```

```python
data["correction_factor"] = mass_coefficient(data.z_median_central)
```
